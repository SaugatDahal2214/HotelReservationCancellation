```python
df = df.drop(columns=['Booking_ID', 'arrival_year'])
```

```python
df.head()
```

```python
df.shape
```

```python
#Distribution of numerical features

df.hist(bins=20, figsize=(20,15))
plt.suptitle("Distribution of numerical feature", fontsize=16)
plt.savefig('numerical_distribution.jpg')
plt.show()
```

```python
# Booking Status Distribution
sns.countplot(x='booking_status', data=df)
plt.title('Booking Status Distribution')
plt.savefig("Booking_distribution.png")
plt.show()
```

```python
#Average price of room based on its type
room_price_avg = df.groupby('room_type_reserved')['avg_price_per_room'].mean().reset_index()

plt.figure(figsize=(10, 10))
sns.barplot(x='room_type_reserved', y='avg_price_per_room', data=room_price_avg)

plt.title('Average Price Of Each Room Type')
plt.xlabel('room_type_reserved')
plt.ylabel('avg_price_per_room')

for i in range(len(room_price_avg)):
    plt.text(i, room_price_avg['avg_price_per_room'][i] + 1, 
             round(room_price_avg['avg_price_per_room'][i], 0), ha='center')

plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.savefig("Correlation_Matrix.jpg")
plt.show()
```

```python
#Categorical features distributions by booking status

categorical_colms = ["type_of_meal_plan", "room_type_reserved", "market_segment_type"]

for col in categorical_colms:
    plt.figure(figsize=(10,10))
    sns.countplot(x=col, hue='booking_status', data=df)
    plt.title(f'{col} Distributed by Booking Status')
    plt.xticks(rotation=45)
    plt.show()
```

```python
#Numeical Features vs Booking Status

numerical_cols = ['lead_time', 'no_of_adults' , 'no_of_children', 'avg_price_per_room', 'no_of_special_requests']

for cols in numerical_cols:
    plt.figure(figsize=(10,10))
    sns.boxplot(x='booking_status', y=cols, data=df)
    plt.title(f'{cols} vs Booking Status')
    plt.show
```

```python
#Distribution of lead time and average price

sns.histplot(df['lead_time'], kde=True)
plt.title('Distribution of lead time')
plt.savefig("Distribution of lead time")
plt.show()

sns.histplot(df['no_of_previous_cancellations'], kde=True)
plt.title('Distribution of Previous Cancellation')
plt.savefig("Distribution of Previous Cancellation")
plt.show()
```

ENCODING CATEGORICAL VALUES

```python
df.head()
```

```python
df['type_of_meal_plan'].unique()
```

```python
#Label Mapping meal type

meal_plan_encoding = {
    'Not Selected':0,
    'Meal Plan 1' : 1,
    'Meal Plan 2' : 2,
    'Meal Plan 3' : 3
}

df['type_of_meal_plan'] = df['type_of_meal_plan'].map(meal_plan_encoding)
```

```python
df['type_of_meal_plan'].unique()
```

```python
df.head
```

```python
df['room_type_reserved'].unique()
```

```python
room_type_encoding = {
    'Room_Type 1': 0,
    'Room_Type 2': 1,
    'Room_Type 3': 2,
    'Room_Type 4': 3,
    'Room_Type 5': 4,
    'Room_Type 6': 5,
    'Room_Type 7': 6,
}

df['room_type_reserved'] = df['room_type_reserved'].map(room_type_encoding)
```

```python
df['room_type_reserved'].unique()
```

```python
df.head()
```

```python
df.select_dtypes(include='object').columns
```

```python
df['market_segment_type'].unique()
```

```python
c1 = len(df[df['market_segment_type'] == 'Corporate'])
c2 = len(df[df['market_segment_type'] == 'Aviation'])
c3 = len(df[df['market_segment_type'] == 'Complementary'])

print(c1, c2, c3)
```

```python
df['market_segment_type'] = df['market_segment_type'].replace(['Aviation', 'Complementary'], 'Corporate')
```

```python
df['market_segment_type'].unique()
```

```python
#Level Encoding Market segment type

market_segment_encoding = {
    'Offline' : 0,
    'Online' : 1,
    'Corporate': 2

}

df['market_segment_type'] = df['market_segment_type'].map(market_segment_encoding)
```

```python
df['market_segment_type'].unique()
```

```python
df.head()
```

```python
df['booking_status'].unique()
```

```python
df.head()
```

```python
#level encoding booking status

Booking_encode ={
    'Not_Canceled':0,
    'Canceled':1
}

df['booking_status'] = df['booking_status'].map(Booking_encode)
```
